```python

import matplotlib
import matplotlib.pyplot as plt

plt.rc('font',family = 'NanumBarunGothic')

plt.title("선 그래프")
plt.plot([4,9,4,2,10])
plt.show()

```

```python
plt.title("선 그래프")
plt.plot([5,10,15,20,25],[4,9,4,2,10],[1,2,2,3,4],c='r',marker='o')
plt.xticks([5,10,15,20,25],['5일','10일','15일','20일','25일'])
plt.xlabel('date')
plt.ylabel('name')
plt.show()
```

```python
plt.title('선그래프')
plt.plot([2,7,3,1],c='g',label="1")
plt.plot([20,70,30,111],c='r',label="2")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(3,2))
plt.bar([2,7,8,11],[1,2,3,4])
plt.show()
```
